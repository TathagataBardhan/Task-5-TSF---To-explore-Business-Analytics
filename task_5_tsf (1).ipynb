#Importing the dataset and the required libraries :

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('/content/drive/My Drive/datasets/SampleSuperstore.csv')
data
```

```
sns.set(style="white")
```

#Visualization of sales with respect to each category:

```
sns.boxplot(x='Category', y='Sales', data=data)
plt.yscale('log')
plt.show()
```

#CONCLUSION : Thus, we can see that sales of products related to office supplies are on an avg. lower than the sales of other two categories.

#Visualization of profits incurred from each category :

```
sns.boxplot(x='Category', y='Profit', data=data)
plt.show()
```

#There are lot of outliers in the data due to which the boxplots are not comprehensible. So, we need to compare sales and profit using any other visualization tool

#Comparison of sales and profit with respect to each category of products :

```
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.barplot(x='Category', y='Sales', data=data, estimator=np.median)
plt.title("Sales")
plt.subplot(1,2,2)
sns.barplot(x='Category', y='Profit', data=data, estimator=np.median)
plt.title("Profit")
plt.show()
```

#CONCLUSION: So, we can see that although the sales of furnitures are high but they incurr least profit compared to office supplies and technology categories. This may be due to the high shipping cost of the furnitures

#So, High sales does not always ensure higher profits. They depend on other factors

#Evaluation of profits incurred with respect to shipping modes and categories :

```
plt.figure(num=None, figsize=(15,8), dpi=80, facecolor='w', edgecolor='k')
sns.barplot(x='Ship Mode', y='Profit', hue='Category', data=data, estimator=np.median)
plt.show()
```

#CONCLUSION : So, we can see that for all types of shipping modes also, prodcuts related to technology incurrs more profits.

#Which shipping mode individually is more profitable :

```
sns.barplot(x='Ship Mode', y='Profit', data=data, estimator=np.median)
plt.show()
```

#CONCLUSION : Thus, all the shipping modes are more or less equal in terms of profit. However, second class ship mode is little better than the other modes in terms of profit

```
plt.figure(num=None, figsize=(15,8), dpi=80, facecolor='w', edgecolor='k')
sns.barplot(x='Segment', y='Profit', hue='Category', data=data, estimator=np.median)
plt.show()
```

#Conclusion : Profit of technology products is maximum for Home office and more than office supplies and furnitures for all customer segments.

#Analysis of sales on product sub-categories :

```
plt.figure(figsize=(20,6))
sns.countplot(y="Sub-Category", data=data)
plt.show()
```

#CONCLUSION : We can see that Binders have maximum sales and Copiers have fewer no. of orders

#N.B.- What is to be noted here is that due to the presence of many outliers in the data, median estimator is used while plotting the barplots.

#Visualizing sales across the regions :

```
plt.figure(num=None, figsize=(15,8), dpi=80, facecolor='w', edgecolor='k')
sns.barplot(x='Region', y='Sales', data=data, estimator=np.median)
plt.show()
```
